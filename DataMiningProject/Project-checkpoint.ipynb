```python
#Encoding(one hot encoding, integer encoding)
#Fonksiyon, "one hot encoding" işlemini gerçekleştirirken, "df" parametresi ile verilen veri kümesini kopyalar ve kopyası üzerinde işlem yapar. "COLUMNS_TO_ENCODE" parametresi ile verilen öznitelikler için "one hot encoding" işlemi gerçekleştirilir ve sonuç olarak, "one hot encoding" işlemine tabi tutulmuş veri kümesi geri döndürülür.
def one_hot_encoding(df, COLUMNS_TO_ENCODE=['Gender', 'Ever_Married','Graduated', 'Profession', 'Var_1']):
    encoded_ds = df.copy()
    for column in COLUMNS_TO_ENCODE:
        previous_index = encoded_ds.columns.get_loc(column)
        one_hot_ds = pd.get_dummies(encoded_ds[column], prefix=column)
        for i, one_hot in enumerate(one_hot_ds):
            encoded_ds.insert(previous_index + i, one_hot, one_hot_ds[one_hot])
        del encoded_ds[column]
    return encoded_ds

def integer_encoding(df, COLUMNS_TO_ENCODE=['Spending_Score']):
    encoded_ds = df.copy()
    for column in COLUMNS_TO_ENCODE:
        previous_index = encoded_ds.columns.get_loc(column)
        integer_encoding = encoded_ds[column].astype('category').cat.codes
        encoded_ds[column] = integer_encoding
    return encoded_ds


encoded_ds = one_hot_encoding(data)
encoded_ds = encoded_ds.drop(columns = ['ID'])
encoded_ds = integer_encoding(encoded_ds)
encoded_ds.head()
```

```python
#Preparing data for classification(scaling)
#COLUMNS_TO_CLEAN" adlı liste "Work_Experience" ve "Family_Size" özniteliklerini içerir. Bu öznitelikler, veri kümesinde eksik 
#değerlerin olup olmadığını kontrol etmek üzere kullanılır.
#"PREDICTED_COLUMN" adlı bir değişken tanımlanır ve bu değişken "Segmentation" özniteliğini içerir. Bu öznitelik,
#veri kümesinde hangi özniteliğin tahmin edileceğini belirtir.
#"scaler" adlı bir değişken tanımlanır ve bu değişkene "StandardScaler" sınıfından bir nesne atanır. 
#"StandardScaler" sınıfı, veri kümesindeki özniteliklerin ortalamasını 0 ve standart sapmasını 1 yapmak için kullanılıyor.
#"prepare_data" adlı bir fonksiyon tanımlanır ve bu fonksiyon "ds" adlı bir parametre alır. Bu parametre, 
#işleme tabi tutulacak olan veri kümesini temsil eder. Fonksiyon içinde, veri kümesinin bir kopyası alınır 
#ve "COLUMNS_TO_CLEAN" listesi içinde yer alan öznitelikler incelenir. Bu özniteliklerin içinde eksik değerler varsa,
#bunlar ortalama değerler ile doldurulur. Daha sonra, veri kümesinde bulunan eksik değerlerin sayıları hesaplanır ve sonuç olarak bu değerlerin sayısı döndürülür.
COLUMNS_TO_CLEAN = ['Work_Experience', 'Family_Size']
PREDICTED_COLUMN = 'Segmentation'
scaler = StandardScaler()
    
def prepare_data(ds, to_fillna=True):
    ds = ds.copy()
    for column in COLUMNS_TO_CLEAN:
        ds[column] = ds[column].fillna(ds[column].mean())
    ds.isna().sum()
    
#"ds" adlı veri kümesinden "PREDICTED_COLUMN" değişkeni olarak belirlenen öznitelik dışındaki öznitelikler bağımsız 
#değişken olarak ("X"), "PREDICTED_COLUMN" değişkeni ise bağımlı değişken olarak ("Y") ayırılır.
#"X_train" ve "X_test" verileri, "StandardScaler" nesnesinin "fit_transform" metodu kullanılarak ölçeklendirilir. 
#Bu ölçeklendirme işlemi, veri kümesindeki özniteliklerin ortalamasını 0 ve standart sapmasını 1 yapar.

    X, Y = ds.drop(PREDICTED_COLUMN, axis=1), ds[PREDICTED_COLUMN]
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=TEST_SIZE, random_state=sklearn_random_state)
    X_train = scaler.fit_transform(X_train.values)
    X_test = scaler.fit_transform(X_test.values)
    #     Y_train = scaler.fit_transform(Y_train.values)
    #     Y_test = scaler.fit_transform(Y_test.values)
    
    return X_train, X_test, Y_train, Y_test

X_train, X_test, Y_train, Y_test = prepare_data(encoded_ds)
```

- SVM algoritması veri setleri üzerinde çok sayıda özellik (feature) olması durumunda da çalışabilir. Ancak bu durumda modelin performansının düşebileceği de unutulmamalıdır.

-> SVM Dezavantajları

- Hiperparametreleri ayarlamak zor olabilir: SVM'de kullanılan hiperparametreler, modelin performansını etkileyebilecek bazı değişkenlerdir. Bu hiperparametrelerin doğru ayarı, modelin performansını iyileştirebilir, ancak yanlış ayarlamak da performansı düşürebilir.

- Veri seti çok büyükse, SVM yavaş çalışabilir: SVM, veri seti büyüdükçe yavaş çalışabilir, bu nedenle çok büyük veri setlerinde kullanımı zor olabilir.

- Veri setlerinde çok sayıda gürültü varsa, SVM performansı düşebilir: SVM, veri setinde çok sayıda gürültü (yani, anlamsız veya yanlış etiketlenmiş veriler) olduğunda performansını düşürebilir.

- SVM, özellikler arasındaki ilişkileri tahmin edemez: SVM, özellikler arasındaki ilişkileri tahmin edemez, bu nedenle özellikler arasındaki ilişkileri anlamaya yönelik çalışmalarda kullanılamaz.

- SVM, sadece iki sınıf problemlerine uygulanabilir: SVM, sadece iki sınıf problemlerine uygulanabilir, bu nedenle çok sınıflı problemler için uygun değildir.

-> Decision Tree Avantajları

- Decision tree algoritması hızlı ve etkili bir öğrenme mekanizmasıdır. Bu, modelin eğitimi ve tahminlerinin yapılması için gereken süreyi azaltır.

- Decision tree algoritması kolayca anlaşılır ve açıklanır. Bir decision tree modelinin yaptığı tahminleri anlamak için, modelin kuruluş sürecini takip etmek yeterlidir. Bu, modelin kuruluş sürecini ve yaptığı tahminleri diğer insanlarla paylaşırken oldukça kolay hale getirir.

- Decision tree algoritması birçok tür veriye uygulanabilir. Bu algoritma sayısal ve kategorik verileri de işleyebilir.

- Decision tree algoritması birçok değişkeni değerlendirebilir. Bir decision tree modeli, birden fazla değişkene göre tahminler yapabilir. Bu, modelin tahminlerinin daha anlamlı olmasını sağlar.

- Decision tree algoritması veri ön işleme gerektirmez. Bu algoritma veri setinin önceden işlenmiş haliyle de çalışabilir.

-> Decision Tree Dezavantajları

- Küçük değişiklikler veri setinde büyük değişikliklere yol açabilir, bu da modelin istikrarsızlığına sebep olabilir.

- Büyük veri setlerinde çok fazla zaman alabilirler.

- Overfitting (aşırı uyum) problemine sahip olabilirler. Yani, eğitim veri setinde iyi sonuçlar verebilirler ancak test veri setinde kötü sonuçlar verebilirler.

- Bazı durumlarda, yüksek öğrenme kapasitesine sahip modele gereksinim duyulabilir. Bu da modelin eğitim süresini uzatabilir.

- Çok fazla özellik olması durumunda, modelin doğru seçim yapması zorlaşabilir.

- Bazı özelliklerin ağırlıklarının çok farklı olması durumunda, modelin performansı düşebilir.

- Decision tree algoritmaları sınıflandırma ve regresyon gibi problemleri çözebilirler, ancak kümeleme gibi problemleri çözemezler.

-> Random Forest Avantajları

- Çok sayıda karar ağacından oluştuğundan, aşırı eğitim ve overfitting problemlerine nadir rastlanır.

- Feature seçimi yapar ve en etkili özellikleri seçerek modellerini optimize eder.

- Verisetinde bir öğe bulunmayan durumları da dikkate alarak tahminlerde bulunur.

- Çok sayıda özellik kullanıldığında bile çalışma hızı yüksektir.

- Birçok tipe uyumludur ve sayısal, kategorik ve ordinal verilere eşit derecede iyi tahminlerde bulunur.


-> Random Forest Dezavantajları

- Modelleme süreci oldukça zaman alıcıdır.

- Anlamaya çalışıldığında oldukça karışık ve anlaşılması zor bir model oluşturur.

- Eğitim sırasında aşırı miktarda RAM kullanımı nedeniyle bazen sistem çökebilir.

- Modelin öğrenme süreci boyunca çok fazla sayıda değişken işlenir ve bu da performansı olumsuz etkileyebilir.

```python
#Random forest classification algorithm

#"RandomForestClassifier" sınıfından bir nesne oluşturulur ve bu nesne "clf" değişkenine atanır. 
#Nesne oluşturulurken, "n_estimators" parametresi olarak 100 değeri verilmiştir. 
#Bu parametre, oluşturulacak ağaçların sayısını belirtir.

#"clf" nesnesinin "fit" metodu kullanılarak, verilen "X_train" ve "Y_train" verilerine göre bir sınıflandırma modeli eğitilir.
#Bu işlem, modelin öğrenme işlemini gerçekleştirir.

#"clf" nesnesinin "predict" metodu kullanılarak, eğitilmiş olan model kullanılarak "X_test" verisi üzerinde tahminler yapılır
#Bu tahminler, "Y_pred" değişkenine atanır.

#"accuracy_score" fonksiyonu kullanılarak, "Y_test" verisi ile yapılan tahminler arasındaki doğruluk oranı hesaplanır. 

clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,Y_train)
Y_pred=clf.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
```

```python
#Finding best paramethers for random forest 

#RandomForestRegressor" sınıfından "rfregr" isimli bir nesne oluşturulur ve bu nesne, "random_state" parametresi olarak verilen
#"sklearn_random_state" değerine göre inşa edilir. Bu parametre, oluşturulacak rastgele ağaçların rastgele seçimleri için
#kullanılan rastgele sayı üretecisinin seed değerini belirtir.

#"n_estimators" anahtarı için bir değer listesi oluşturulur. Bu liste, "start" ve "stop" parametreleri arasındaki sayılardan,
#"num" parametresi kadar eşit aralıklı sayıları içerir. Örneğin, "start=200", "stop=2000", "num=10" olduğu için, bu liste 200,
#400, 600, ..., 2000 değerlerini içerecektir.

#"max_features" anahtarı için bir değer listesi oluşturulur. Bu liste, "auto" ve "sqrt" değerlerini içerir.
#"max_depth" anahtarı için bir değer listesi oluşturulur. Bu liste, "10" ve "110" arasındaki sayılardan, "11" adet 
#eşit aralıklı sayıyı içerir. Bu listeye ayrıca "None" değeri eklenir.

#"bootstrap" anahtarı için bir değer listesi oluşturulur. Bu liste, "True" ve "False" değerlerini içerir.
#"min_samples_split" anahtarı için bir değer listesi oluşturulur. Bu liste, 2, 5 ve 10 değerlerini içerir.

#"random_grid" adında bir dictionary oluşturulur. Bu dictionary'de, önceki adımda oluşturulan değer listelerini anahtar 
#olarak kullanılır. Bu şekilde, öğrenme hiperparametreleri için bir arama grubu oluşturulmuş olur.

#Bu arama grubu, daha sonra kullanılacak bir "RandomizedSearchCV" nesnesi tarafından kullanılarak,
#en uygun hiperparametrelerin belirlenmesi için kullanılır.

rfregr = RandomForestRegressor(random_state = sklearn_random_state)

n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
#Fitting for best paramethers

#"RandomizedSearchCV" sınıfından bir nesne oluşturulur ve bu nesne "rf_random" değişkenine atanır.

#Bu nesne, verilen bir veri kümesine göre bir modelin hiperparametrelerinin belirlenmesi için kullanılır. 

#Nesne oluşturulurken, "estimator" parametresi olarak "rfregr" nesnesi, "param_distributions" parametresi olarak 
#"random_grid" dictionary'si, "n_iter" parametresi olarak 100, "cv" parametresi olarak 3, "verbose" parametresi olarak 2 
#ve "random_state" parametresi olarak 42 değerleri verilmiştir. "n_jobs" parametresi olarak "-1" değeri verilmiştir.
#Bu parametre, modelin eğitim sırasında kullanılacak iş parçacıklarının sayısını belirtir.

#"rf_random" nesnesinin "fit" metodu kullanılarak, verilen "X_train" ve "y_train_labels" verilerine göre bir 
#"RandomForestRegressor" modeli eğitilir ve bu modelin hiperparametreleri belirlenir. Bu işlem, verilen hiperparametre 
#arama grubunu kullanarak, en uygun hiperparametrelerin belirlenmesi için bir arama işlemi gerçekleştirir.

#"rf_random" nesnesinin "best_params_" özelliği kullanılarak, en uygun hiperparametrelerin bulunduğu bir dictionary döndürülür.
#Bu dictionary, en uygun hiperparametreleri ve bu hiperparametrelere ait değerleri içerir. 
#Bu değerler, önceki adımda oluşturulan "random_grid" arama grubunda belirtilen değerler arasından seçilmiştir.

rf_random = RandomizedSearchCV(estimator = rfregr, param_distributions = random_grid,
                               n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(X_train, y_train_labels)
rf_random.best_params_
```

```python
#Decision tree classification algorithm

#DecisionTreeClassifier sınıfından bir nesne oluşturulur ve bu nesne "dt" değişkenine atanır. Bu nesne, "max_depth" 
#parametresi olarak 3 değerine göre inşa edilir.

#dt nesnesinin fit metodu kullanılarak, verilen "X_train" ve "y_train_labels" verilerine göre bir "DecisionTreeClassifier" 
#modeli eğitilir.

#dt nesnesinin predict metodu kullanılarak, eğitilen model kullanılarak "X_train" ve "X_test" verilerine ait tahminler 
#yapılır ve bu tahminler "y_train_pred" ve "y_test_pred" değişkenlerine atanır.

#accuracy_score fonksiyonu kullanılarak, "y_train_labels" ve "y_train_pred" verileri arasındaki doğruluk oranı 
#hesaplanır ve ekrana yazdırılır.

#confusion_matrix fonksiyonu kullanılarak, "y_train_labels" ve "y_train_pred" verileri arasındaki karışıklık matrisi hesaplanır ve bu matris döndürülür.


dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train, y_train_labels)

y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

print(accuracy_score(y_train_labels, y_train_pred))
confusion_matrix(y_train_labels, y_train_pred)
```

```python
#Fitting for best paramethers 
#params adlı bir dictionary oluşturulur. Bu dictionary, "DecisionTreeClassifier" modelinin hiperparametrelerini ve bu hiperparametrelerin
#alabileceği değerleri içerir.Max_depth hiperparametresinin alabileceği değerler 2, 3, 5, 10, 20, 50 ve 100'dür.

#"GridSearchCV" sınıfından bir nesne oluşturulur ve bu nesne "grid_search" değişkenine atanır. Bu nesne, 
#"DecisionTreeClassifier" modelini, verilen hiperparametreleri ve değerlerini kullanarak, en uygun hiperparametrelerin 
#belirlenmesi için bir arama işlemi gerçekleştirir.

#"grid_search" nesnesinin "fit" metodu kullanılarak, verilen "X_train" ve "y_train_labels" verilerine göre bir
#"DecisionTreeClassifier" modeli eğitilir ve bu modelin hiperparametreleri belirlenir.

#"grid_search" nesnesinin "best_params_" özelliği kullanılarak, en uygun hiperparametrelerin bulunduğu bir dictionary döndürülür.
params = {
    'max_depth': [2, 3, 5, 10, 20, 50, 100],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'criterion': ["gini", "entropy"]
}

grid_search = GridSearchCV(estimator=dt, 
                           param_grid=params, 
                           cv=4, n_jobs=-1, verbose=1, scoring = "accuracy")
grid_search.fit(X_train, y_train_labels)
grid_search.best_params_
```

```python
#SVM classification algorithm
#"svm.SVC" sınıfından bir nesne oluşturulur ve bu nesne "linear" değişkenine atanır. Bu nesne, "linear" kernel fonksiyonunu, 
#"ovo" (One-vs-One) sınıflandırma yöntemini ve "C" parametresinin 1 değerine göre inşa edilir. Daha sonra bu nesnenin "fit"
#metodu kullanılarak, verilen "X_train" ve "Y_train" verilerine göre bir "Support Vector Machine" modeli eğitilir.

#"svm.SVC" sınıfından bir nesne oluşturulur ve bu nesne "rbf" değişkenine atanır. Bu nesne, "radial basis function" kernel 
#fonksiyonunu, "ovo" sınıflandırma yöntemini ve "C" parametresinin 1 değerine göre inşa edilir. "gamma" parametresine de 0.1 
#değeri verilir. Daha sonra bu nesnenin "fit" metodu kullanılarak, verilen "X_train" ve "Y_train" verilerine göre bir "Support Vector Machine" modeli eğitilir.

#"svm.SVC" sınıfından bir nesne oluşturulur ve bu nesne "sig" değişkenine atanır. Bu nesne, "sigmoid" kernel fonksiyonunu, 
#"ovo" sınıflandırma yöntemini ve "C" parametresinin 1 değerine göre inşa edilir. Daha sonra bu nesnenin "fit" metodu 
#kullanılarak, verilen "X_train" ve "Y_train" verilerine göre bir "Support Vector Machine" modeli eğitilir.

#"kernels", "methods" ve "names" adlı dört adet liste oluşturulur. Bu listeler, sırasıyla, oluşturulan dört 
#"Support Vector Machine" modelinin kernel fonksiyonlarını, bu modelleri ve bu modellerin isimlerini içerir.
#Bu listeler, daha sonra kullanılacak olan kod parçacıklarında referans olarak kullanılacaktır.


linear = svm.SVC(kernel='linear', C=1,
                decision_function_shape='ovo').fit(X_train, Y_train)
rbf = svm.SVC(kernel='rbf', gamma=0.1, C=1,
              decision_function_shape='ovo').fit(X_train, Y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1,
               decision_function_shape='ovo').fit(X_train, Y_train)
sig = svm.SVC(kernel='sigmoid', C=1,
              decision_function_shape='ovo').fit(X_train, Y_train)

kernels = ['linear', 'rbf', 'poly', 'sigmoid']
methods = [linear, rbf, poly, sig]
names = ['Linear', 'Radial Basis', 'Polynomial', 'Sigmoid']
```

```python
#Daha önce oluşturulan dört "Support Vector Machine" modelinin doğruluğu ölçülür.

#"enumerate" fonksiyonu kullanılarak, "methods" listesinin elemanları (yani oluşturulan "Support Vector Machine" modelleri)
#tek tek döndürülür. Bu döngüde, her bir model için "i" değişkenine modelin sıra numarası, "method" değişkenine ise modelin 
#nesnesi atanır.

#"method" değişkeninin "predict" metodu kullanılarak, "X_test" verilerine göre tahminler yapılır ve bu tahminler 
#"pred" değişkenine atanır.

#"method" değişkeninin "score" metodu kullanılarak, "X_test" ve "Y_test" verilerine göre modelin doğruluğu ölçülür ve 
#"accuracy" değişkenine atanır.

"f-strings" kullanılarak ekrana yazdırılır. Bu yazdırma işleminde, modelin sıra numarasına göre "names" listesinden alınan isim kullanılır ve modelin doğruluğu yüzdelik bir oran olarak yazdırılır.
for i, method in enumerate(methods):
    pred = method.predict(X_test)
    accuracy = method.score(X_test, Y_test)
    print(f'Accuracy {names[i]} Kernel: {accuracy:.2f}')
```

```python
#Finding best paramethers for SVM
#"tuned_parameters" değişkenine, SVM modeli için farklı kernel tiplerine ve bu kernel tiplerine uygun farklı parametrelerine
#ait bir liste atanır. Bu listede, her bir kernel tipi için hangi parametrelerin deneyeceği belirtilir.

#"enumerate" fonksiyonu kullanılarak, "kernels" listesinin elemanları (yani "linear", "rbf", "poly" ve "sigmoid" kernel tipleri)
#tek tek döndürülür. Bu döngüde, her bir kernel için "i" değişkenine kernelin sıra numarası, "kernel" değişkenine ise kernelin 
#ismi atanır.

#"GridSearchCV" sınıfı kullanılarak, "tuned_parameters" listesinde belirtilen kernel tipleri ve parametreleri deneyen 
#bir model oluşturulur. Bu modele ait nesne "svm.grid" değişkenine atanır.

#"svm.grid" nesnesinin "fit" metodu kullanılarak, "X_train" ve "Y_train" verilerine göre model eğitilir.

#"svm.grid" nesnesinin "best_params_" özelliği kullanılarak, en iyi performansı gösteren parametreler ekrana yazdırılır.

tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4, 1, 1e-1, 2, 4, 5],
                     'C': (0.1,1,10,100,1000)},
                     {'kernel': ['linear'], 'C': (0.1,1,10,100,1000)},
                     {'kernel': ['poly'], 'degree': [1,2,3,4,5,6,7,8],
                     'C': (0.1,1,10,100,1000)},
                     {'kernel': ['sigmoid'], 'C': (0.1,1,10,100,1000)}]
for i, kernel in enumerate(kernels):
    svm.grid = GridSearchCV(svm.SVC(), tuned_parameters,
                            n_jobs=-1, cv=10, verbose=1, scoring='accuracy')
svm.grid.fit(X_train, Y_train)
print(svm.grid.best_params_)
```

```python
#Finding best SVM method
#"rbf" kernel tipi ve "gamma" ve "C" parametreleri kullanılarak bir SVM modeli oluşturulur. 
#Bu model "X_train" ve "Y_train" verilerine göre eğitilir ve "best_SVM_method" değişkenine atanır.

#"best_SVM_method" modeli kullanılarak "X_test" verisindeki gözlemlerin sınıflandırılması yapılır ve bu tahminler 
#"pred" değişkenine atanır.

#"best_SVM_method" modelinin "X_test" ve "Y_test" verilerine göre doğruluk oranı (accuracy) hesaplanır ve ekrana yazdırılır.

best_SVM_method = rbf = svm.SVC(kernel='rbf', gamma=0.1, C=1,).fit(X_train, Y_train)
pred = best_SVM_method.predict(X_test)
accuracy = best_SVM_method.score(X_test, Y_test)
print(f'Accuracy Kernel: {accuracy:.2f}')
```
